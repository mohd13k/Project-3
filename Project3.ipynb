```python
# Project 3: Object Masking and Component Detection
# Mohammad Khan
# 500976279

import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
import shutil
import os

# Step 1: Object Masking
image_path = "/content/drive/MyDrive/850 Project 3/motherboard_image.JPEG"
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

blurred = cv2.GaussianBlur(gray, (9, 9), 0)
adaptive_thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 19, 5)
edges = cv2.Canny(adaptive_thresh, 30, 150)
dilated_edges = cv2.morphologyEx(edges, cv2.MORPH_DILATE, np.ones((5, 5), np.uint8))

contours, _ = cv2.findContours(dilated_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

if contours:
    largest_contour = max(contours, key=cv2.contourArea)
    edge_with_contour = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)
    cv2.drawContours(edge_with_contour, [largest_contour], -1, (255, 255, 255), 5)
    mask = np.zeros_like(gray)
    cv2.drawContours(mask, [largest_contour], -1, 255, -1)
    extracted_pcb = cv2.bitwise_and(image, image, mask=mask)
    extracted_path = "/content/drive/MyDrive/850 Project 3/extracted_pcb_clean.JPEG"
    cv2.imwrite(extracted_path, extracted_pcb)
    fig, ax = plt.subplots(1, 4, figsize=(20, 5))
    ax[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    ax[0].set_title("Original Image")
    ax[0].axis("off")

    ax[1].imshow(cv2.cvtColor(edge_with_contour, cv2.COLOR_BGR2RGB))
    ax[1].set_title("Edge Detection with Contour")
    ax[1].axis("off")

    ax[2].imshow(mask, cmap='gray')
    ax[2].set_title("Mask")
    ax[2].axis("off")

    ax[3].imshow(cv2.cvtColor(extracted_pcb, cv2.COLOR_BGR2RGB))
    ax[3].set_title("Extracted PCB")
    ax[3].axis("off")

    plt.tight_layout()
    plt.show()
else:
    print("No contours found")

# Step 2: YOLOv8 Training
data_yaml_path = "/content/drive/MyDrive/850 Project 3/data.yaml"
output_model_name = "pcb_detection"
trained_model_path = f"/content/drive/MyDrive/850 Project 3/{output_model_name}.pt"

model = YOLO("yolov8n.pt")

# Train the YOLOv8 model
model.train(
    data=data_yaml_path,
    epochs=100,
    batch=16,
    imgsz=704,
    name=output_model_name
)

# Save the trained model
runs_dir = "runs/detect"
latest_folder = max([os.path.join(runs_dir, d) for d in os.listdir(runs_dir)], key=os.path.getmtime)
source_model_path = os.path.join(latest_folder, "weights/best.pt")

if not os.path.exists(source_model_path):
    raise FileNotFoundError(f"Trained model not found at {source_model_path}. Check the training output.")

shutil.copy(source_model_path, trained_model_path)
print(f"Trained model saved at: {trained_model_path}")

# Step 3: YOLOv8 Evaluation
evaluation_folder = "/content/drive/MyDrive/850 Project 3/evaluation"
output_folder = "/content/drive/MyDrive/850 Project 3/evaluation_results"

os.makedirs(output_folder, exist_ok=True)

# Evaluates the model on images in the evaluation folder
for image_file in os.listdir(evaluation_folder):
    image_path = os.path.join(evaluation_folder, image_file)
    if image_file.lower().endswith(('.png', '.jpg', '.jpeg')):
        print(f"Evaluating on {image_file}...")
        results = model.predict(
            source=image_path,
            conf=0.25,
            save=True,
            save_txt=True,
            project=output_folder,
            name=image_file.split('.')[0]
        )
        print(f"Results saved for {image_file}.")

print(f"Evaluation results saved in: {output_folder}")
```
